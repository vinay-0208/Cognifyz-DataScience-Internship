# Task 2: Descriptive Analysis

```python
df["Country Code"].value_counts()
```

```python
df["City"].value_counts().head(10)
```

```python
df["Cuisines"].value_counts().head(10)
```

```python
top_cities = df["City"].value_counts().head(10)

plt.figure(figsize=(10,5))
top_cities.plot(kind='bar')
plt.title("Top 10 Cities with Most Restaurants")
plt.xlabel("City")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=45)
plt.show()
```

### Observation

- Certain cities contain a significantly larger number of restaurants.
- Restaurant distribution is concentrated in major urban areas.
- These cities dominate the dataset.

```python
top_cuisines = df["Cuisines"].value_counts().head(10)

plt.figure(figsize=(10,5))
top_cuisines.plot(kind='bar')
plt.title("Top 10 Cuisines")
plt.xlabel("Cuisine")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

### Observation

- A few cuisines appear much more frequently than others.
- Popular cuisines dominate the restaurant market.
- Cuisine diversity is visible throughout the dataset.

#Task 3 — Geospatial Analysis

```python
import folium
```

```python
df[["Latitude", "Longitude"]].head()
```

```python
restaurant_map = folium.Map(
    location=[df["Latitude"].mean(), df["Longitude"].mean()],
    zoom_start=5
)
restaurant_map
```

### Observation

- Restaurants are distributed across multiple geographical locations.
- The map visualization helps identify areas with high restaurant density.
- Restaurant clusters are visible in urban regions.

```python
for index, row in df.head(100).iterrows():
    folium.Marker(
        [row["Latitude"], row["Longitude"]],
        popup=row["Restaurant Name"]
    ).add_to(restaurant_map)

restaurant_map
```

### Observation

- Restaurants are distributed across multiple geographical locations.
- The map visualization helps identify areas with high restaurant density.
- Restaurant clusters are visible in urban regions.

```python
city_counts = df["City"].value_counts().head(10)

plt.figure(figsize=(10,5))
city_counts.plot(kind='bar')
plt.title("Restaurant Distribution by City")
plt.xlabel("City")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

### Observation

- Restaurant distribution varies considerably across cities.
- Some cities contribute a large share of restaurants in the dataset.
- Geographic concentration is evident.

```python
plt.figure(figsize=(8,6))

plt.scatter(
    df["Longitude"],
    df["Latitude"],
    c=df["Aggregate rating"],
    cmap="viridis"
)

plt.colorbar(label="Aggregate Rating")

plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Restaurant Locations vs Ratings")

plt.show()
```

### Observation

- Restaurant ratings vary across geographical locations.
- No strong visual pattern is immediately apparent between location and rating.
- Additional analysis would be required to determine significant geographical effects.
